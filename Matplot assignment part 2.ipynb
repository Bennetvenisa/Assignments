10. In the diabetes dataset, plot a pie-chart showing % of diabetes and % of non-diabetes.

```python
diabetes=pd.read_csv('D:\Datascience\datas\diabetes.csv')
```

```python
diabetes.head()
```

```python
diabetes.describe()
```

```python
diabetes['Outcome'].value_counts()
```

```python
labels=['Diabetes', 'No_Diabetes']
```

```python
plt.figure(figsize=(8,10))
size=[diabetes.Outcome[diabetes['Outcome']==1].count(),diabetes.Outcome[diabetes['Outcome']==0].count()]
plt.pie(size,explode=(0,0.1), labels=labels, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```

6.In the breast cancer dataset, visualize the malignant and benign diagnosis

```python
cancer=pd.read_csv('D:\\Datascience\\datas\\breast_cancer.csv')
```

```python
cancer.head()
```

```python
cancer['malignant'].describe().T
```

```python
 cancer['benign'].describe().T
```

```python
cancer['benign'].value_counts()
```

```python
group=cancer.groupby('benign')
```

```python
benign_pos=group.get_group(1)
benign_pos.head()
```

```python
benign_neg=group.get_group(0)
benign_neg.head()
```

```python
plt.figure(figsize=(15,10))
ax1=plt.subplot(1,2,1)
ax1=sns.distplot(benign_pos['malignant'])
ax1.title.set_text('Maglignant in Benign Positive patients')

ax2=plt.subplot(1,2,2)
ax2=sns.distplot(benign_neg['malignant'])
ax2.title.set_text('Maglignant in Benign Negative patients')
```

```python
plt.figure(figsize=(8,15))
ax1=plt.subplot(2,1,1)
ax1=sns.distplot(cancer['malignant'])
ax1.title.set_text('Malignant')

ax2=plt.subplot(2,1,2)
size=[cancer.benign[cancer['benign']==1].count(), cancer.benign[cancer['benign']==0].count()]
labels=['Benign_positive', 'Benign_negative']
explode=(0,0.1)
ax2=plt.pie(size,explode,labels=labels,autopct='%1.1ff%%')
```

7. In the same breast cancer dataset, plot the columns 'radius_mean', 'texture_mean' along with the col 'diagnosis'.

```python
can_df=pd.read_csv('D:\\Datascience\\datas\\breast-cancer-data.csv')
can_df.head()
```

```python
can_df.describe().T
```

```python
can_df['diagnosis']=can_df['diagnosis'].map({'B':1, 'M':0})
```

```python
can_df['diagnosis'].value_counts()
```

```python
plt.figure(figsize=(10,8))
w=0.3
#ax=plt.subplot(111)
plt.bar(can_df['diagnosis']+w,can_df['radius_mean'], width=0.3, color='b', align='center')
plt.bar(can_df['diagnosis'],can_df['texture_mean'], width=0.3, color='g', align='center')
plt.xticks(np.arange(2),('-ve diagnosis', '+ve diagnosis'))
plt.legend(('Radius_mean', 'Texture_mean'))
#ax.set_xticklabels
plt.show()
```
