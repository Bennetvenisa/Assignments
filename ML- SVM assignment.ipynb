4.Use statistical test , apart from heatmap to get the collinearity among features


Chi-Square test

```python
from scipy.stats import chi2_contingency
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
numeric=forest.loc[:,num_col]
nums=MinMaxScaler().fit_transform(numeric)
cont=pd.DataFrame(nums,columns=cont.columns)
```

```python
forest=pd.concat([cont,category], axis=1)
forest.shape
```

```python
forest.head ()
```

```python
p_values=[]
for col in forest.columns:
    cont=pd.crosstab(forest[col],forest['Cover_Type'])
    chi2_stats, p_val, dof, table=chi2_contingency(cont)
    p_values.append(p_val)
print(p_values)  
```

```python
p_values=np.round(p_values,4)
```

```python
feature_stats=pd.DataFrame({'feature':forest.columns, 'p_values':p_values})
feature_stats
```

variables:soil_type25, soil_type15,soil_type8, soil_type7 show no relationship the target and itself. remaining all other variables contribute to the prediction of the forest types
