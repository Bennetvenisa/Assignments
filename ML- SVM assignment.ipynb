4.Use statistical test , apart from heatmap to get the collinearity among features


Chi-Square test

```python
from scipy.stats import chi2_contingency
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
numeric=forest.loc[:,num_col]
nums=MinMaxScaler().fit_transform(numeric)
cont=pd.DataFrame(nums,columns=cont.columns)
```

```python
forest=pd.concat([cont,category], axis=1)
forest.shape
```

```python
forest.head ()
```

```python
p_values=[]
for col in forest.columns:
    cont=pd.crosstab(forest[col],forest['Cover_Type'])
    chi2_stats, p_val, dof, table=chi2_contingency(cont)
    p_values.append(p_val)
print(p_values)  
```

```python
p_values=np.round(p_values,4)
```

```python
feature_stats=pd.DataFrame({'feature':forest.columns, 'p_values':p_values})
feature_stats
```

variables:soil_type25, soil_type15,soil_type8, soil_type7 show no relationship the target and itself. remaining all other variables contribute to the prediction of the forest types

6.Make use of PCA in reducing the dimensionality

```python
x=forest.iloc[:,0:-1]
y=forest['Cover_Type']
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=5)
```

```python
comp=pca.fit_transform(x)
principal_df=pd.DataFrame(comp,columns=('PC1','PC2', 'PC3','PC4','PC5'))
```

```python
principal_df.head()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
(x_train,x_test,y_train,y_test)=train_test_split(principal_df,y, test_size=0.2, random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

8.Verify the effect of C parameter and gamma function. Plot the accuracy of train and test score

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
from sklearn.svm import SVC,LinearSVC
```

```python
svc=SVC()
```

```python
model=svc.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
C_range = list(np.arange(0.1,2,0.1))
acc_score = []
for c in C_range:
    svc = SVC(kernel='linear',C=c)
    scores = cross_val_score(svc, x_train, y_train,scoring='accuracy')
    acc_score.append(scores.mean())
print(acc_score)
```

```python
plt.plot(c_range,acc_score)
plt.xlabel('c-parameter')
plt.ylabel('Accuracy score')
plt.title('accuracy score for different c parameter')
```

```python
gamma=[0.0001,0.001,0.01,0.1]
```

```python
acc_score =[]
for g in gamma:
    svc = SVC(kernel='rbf', gamma=g)
    scores = cross_val_score(svc, x_train, y_train, cv=10, scoring='accuracy')
    acc_score.append(scores.mean())
print(acc_score)
```

```python
plt.plot(gamma,acc_score)
plt.xlabel('Gamma')
plt.ylabel('Accuracy score')
plt.title('accuracy score for different Gamma')
```

```python
from sklearn import GridSearchCV
```

```python
param_grid = {'C': C_range, 'gamma' : gamma}
grid_search = GridSearchCV(SVC(kernel='rbf'), param_grid, cv=10)
grid_search.fit(x_train, y_train)
grid_search.best_params_
```
