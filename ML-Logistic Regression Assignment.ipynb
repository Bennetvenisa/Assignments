4.Perform Feature Enginnering, atleast 2-3 features, Normalize features using relevant scaling techniques


```python
#Lets scale the features
num_col=[]
for col in data.select_dtypes('float'):
    num_col.append(col)
```

```python
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
nums=data.loc[:,num_col]
numeric=MinMaxScaler().fit_transform(nums)
num_df=pd.DataFrame(numeric,columns=num_col)
```

```python
le=LabelEncoder()
```

```python
data['diagnosis']=le.fit_transform(data['diagnosis'].astype('str'))
```

```python
cat_df=data['diagnosis']
```

```python
data=pd.concat([num_df,cat_df], axis=1)
data.head()
```

Check for the p-value in features

```python
from scipy.stats import chisquare, chi2_contingency
```

```python
chisquare(data['texture_mean'].value_counts())
```

```python
p_values=[]
for col in num_col:
    cont=pd.crosstab(data['diagnosis'],data[col])
    chi2_stats,p_val,dof, tabel=chi2_contingency(cont)
    p_values.append(p_val)
```

```python
PValues=pd.DataFrame({'columns':num_col, 'p_values':p_values})
PValues
```

from the above p-values we can see that variables like-- 'Concave_points_worst, Radius_worst, Radisu_mean... are the significant variables for the diagnosis 

#### Modeling: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
```

```python
x=data.drop('diagnosis',  axis=1)
```

```python
y=data['diagnosis']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
logreg=LogisticRegression()
model=logreg.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
result=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
result.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix=confusion_matrix(y_test,y_pred)
confusion_matrix
```

```python
model_accuracy=((37+72)/(37+72+0+5))*100
print('Model acccuracy is:', model_accuracy,'%')
```

```python
result=classification_report(y_test,y_pred)
result
```

```python
from sklearn.metrics import roc_curve,auc
```

```python
fpr,tpr,thresholds=roc_curve(y_test,y_pred)
roc_auc=auc(fpr,tpr)
```

```python
tpr
```

```python
print('True Positive Rate:', tpr)
print('False positive Rate',fpr)
print('Area under ROC curve:%f'% roc_auc)
```
